```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/processed/combined_acc_walking.csv")
```

```python
# Display the first few rows of the dataset
print("Dataset Preview:")
display(data.head())
```

```python
# Check unique sources
print("Sources in Dataset:", data['source'].unique())
```

```python
# Filter data for specific sensors
chest_data = data[data['source'] == 'chest']
waist_data = data[data['source'] == 'waist']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(chest_data['attr_time'], chest_data['attr_x'], label='Chest X-axis')
plt.plot(waist_data['attr_time'], waist_data['attr_x'], label='Waist X-axis')
plt.legend()
plt.title('Acceleration Data Over Time (X-axis)')
plt.xlabel('Time (ms)')
plt.ylabel('Acceleration')
plt.show()
```

```python
# Group by source and calculate correlations for each sensor
for source in data['source'].unique():
    sensor_data = data[data['source'] == source][['attr_x', 'attr_y', 'attr_z']]
    print(f"Correlation Heatmap for {source.capitalize()} Sensor")
    sns.heatmap(sensor_data.corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title(f"{source.capitalize()} Sensor Correlation")
    plt.show()
```
